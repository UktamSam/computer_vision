```python
import  matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])

plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
from matplotlib import lines
x = np.array([1, 2, 3, 4])
y = np.array([3, 8, 1, 10])

plt.plot(y, linestyle = "dotted", color="green", marker="o")
plt.title("Styled Line Example")
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([2, 5, 7, 8])

plt.plot(x, y1, label = "line1", color="red", marker="s")
plt.plot(x, y2, label = "line2", color="blue", linestyle="--", marker="o")
plt.title("Multiple Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

**Bar Chart with Numpy and Matplotlib**

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([5, 9, 6, 4])

plt.bar(x, y, color="orange")
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

**Scatter Plot with Numpy and Matplotlib**

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 7])

plt.scatter(x, y, color="purple", s=100)
plt.title("Scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

**Histogram with Numpy and Matplotlib**

```python
data = np.array([3, 4, 5, 6, 8, 8, 8, 9, 10])

plt.hist(data, bins=5, color="skyblue", edgecolor="black")
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Counts")
plt.show()
```

**Subplots(Multiple Graphs Together**

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([2, 5, 7, 8])

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y1, color="red")
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.bar(x, y2, color="green")
plt.title("Plot 2")

plt.tight_layout()
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([4, 8, 6, 10])

plt.plot(x, y, color="green")
plt.title("Save Figure Example")
plt.savefig("myplot.png") # Saves in current folder
plt.show()
```

```python
plt.style.use('_mpl-gallery-nogrid')

# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
levels = np.linspace(Z.min(), Z.max(), 7)

# plot
fig, ax = plt.subplots()

ax.contourf(X, Y, Z, levels=levels)

plt.show()
```

```python
plt.style.use('_mpl-gallery')

# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap="Blues")

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
